```python
import xml.etree.ElementTree as ET
import requests
import json
import csv
```

```python
def save_xml(url):
    response = requests.get(url)
    with open("movies.xml", "wb") as file:
        file.write(response.content)
save_xml("https://news.google.com/rss/search?q=sport")
```

```python
tree = ET.parse("sports.xml")
root = tree.getroot()
print("root:")
print("tags:", root.tag, "\t attribte ", root.attrib )
print("elements:")
for element in root:
    for sub_element in element:
        #print attributes
        print("tag:", sub_element.tag, "\t\t", "attribute", sub_element.attrib)
        #print the text
        print(sub_element, ":", sub_element.text)
```

```python

def parse_xml(file):
    tree = ET.parse(file)
    root = tree.getroot()
    title_list = []
    for element in root.findall("./channel/item"):
        for sub_element in element:
            if sub_element.tag == "title":
                title_list.append(sub_element.text)
    return title_list

title_list = parse_xml("sports.xml")
for i in range(len(title_list)):
    print(f"News Title {i}: {title_list[i]}")
```

```python
#function for loading json files
def load_json(file_name):
    with open(file_name) as file:
        json_data = json.load(file)
    return json_data

#function for writing json data
json_obj = {
    "name":"larry",
    "age":20,
    "friends":("John", "Sarah")
}
def write_json(json_obj, file_name):
    with open(file_name, "w") as file:
        json.dump(json_obj, file)
write_json(json_obj, "example.json")  
```

```python
#only save completed tasks
def save_completed_tasks(url):
    response = requests.get(url)
    input_list = json.loads(response.text)
    completed_list = []
    for i in input_list:
        if i["completed"] == True:
            completed_list.append(i)
    return json.dumps(completed_list)

print(save_completed_tasks("https://cutt.ly/json"))
```
